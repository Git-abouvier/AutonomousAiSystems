```python
import time
from colorama import Fore
def print_text_animated(text):
    for char in text:
        print(char, end="", flush=True)
        time.sleep(0.02)
```

```python
from camel.agent import RolePlaying
  
task_prompt = "Help me create a html website. This website is about the Chrome extension called 'Mi-GPT'. MiGPT Chrome extension is designed to improve your writing and reading experience with a suite of powerful features, powered by GPT-4 AI. With just a few clicks, you can access the following features: Capitalize Title: Automatically capitalize selected text according to title case standards. Write an Outline: Generate a blog outline based on the selected text. Write a Paragraph: Create a paragraph on a given topic or prompt. Fix Grammar: Correct grammar mistakes in the selected text. Summarize Text: Generate a concise summary of the selected text. Add Call-to-Action (CTA): Insert a customizable call-to-action in your content. Ask AI: Get answers or suggestions from GPT-4 based on a selected question or prompt. GPT-4 Prompts: Access a database of GPT-4 prompts to explore new ideas. Share on Twitter: Share selected text directly to Twitter. Exclude Page: Blacklist specific pages from displaying the toolbar. MiGPT also offers support for Medium writers with additional features tailored to Medium's editor. The extension is easy to use, customizable, and can be toggled on or off as needed."
print(Fore.YELLOW + f"Original task prompt:\n{task_prompt}\n")
role_play_session = RolePlaying("Web Developer", "Web Developer", task_prompt)
print(Fore.CYAN + f"Specified task prompt:\n{role_play_session.task_prompt}\n")

chat_turn_limit, n = 10, 0
assistant_msg, _ = role_play_session.init_chat()
while n < chat_turn_limit:
    n += 1
    (assistant_msg, _, _), (user_msg, _, _) = role_play_session.step(assistant_msg)
    print_text_animated(Fore.BLUE + f"AI User:\n\n{user_msg.content}\n\n")
    print_text_animated(Fore.GREEN + f"AI Assistant:\n\n{assistant_msg.content}\n\n")
    if "<CAMEL_TASK_DONE>" in assistant_msg.content:
        break
```
